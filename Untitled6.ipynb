##import dependences

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import sklearn.datasets
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

##Data Collection and processing

```python
breast_cancer_dataset = sklearn.datasets.load_breast_cancer()
```

```python
print(breast_cancer_dataset)
```

```python
data = pd.DataFrame(breast_cancer_dataset.data,columns=breast_cancer_dataset.feature_names)
```

```python
data['label'] = breast_cancer_dataset.target
```

```python
x = data.drop(columns='label',axis=1)
y = data['label']
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=2)
```

```python
LOG_model = LogisticRegression()
```

```python
LOG_model.fit(x_train,y_train)
```

```python
y_pred = LOG_model.predict(x_test)
```

```python
accuracy_score(y_test,y_pred)
```

#Bulidding ANN model

```python
import tensorflow
from tensorflow import keras
from tensorflow.keras import Sequential
from tensorflow.keras.layers import Dense
```

```python
x.shape
```

```python
keras.losses.MeanAbsoluteError
```

```python
def model_bulding(hp):
  model = Sequential()

  units = hp.Int('units',min_value = 1,max_value =30 ,step = 5)
  activation = hp.Choice('activtion',values=['relu','sigmoid',"tanh"])
  optimizer = hp.Choice('optimizer',values = ['Adam','rmsprop'])

  model.add(Dense(units=units,activation=activation,input_dim=30))
  model.add(Dense(1,activation=activation))

  model.compile(optimizer=optimizer,loss = keras.losses.MeanAbsoluteError,metrics=['accuracy'])

  return model
```
